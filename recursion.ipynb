```python
# calling a function itself
 #means calling a function inside a function
# bydefault we can do this recursion only for 3000 times
def greet():
    print("hello")
    greet()          #calling a function inside the function so the function will goes calling
greet()
```

```python
# to know the limit of recursion
 # we can check it by importing a sys function and than calling the method
import sys
print(sys.getrecursionlimit())   # here it is showing the limit of recursion
```

```python
# to set the recursion limit
#that we can do by calling a metho 
import sys
sys.setrecursionlimit(100)
```

```python
import sys
sys.setrecursionlimit(100)
count=0
def greet():
    global count
    count=count+1
    
    print("hello",count)
    greet()          #calling a function inside the function so the function will goes calling
greet()
```
